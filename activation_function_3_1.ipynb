```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

### sigmoid

```python
x = np.linspace(-10, 10, 100)
y1 = 1/(1+(math.e)**(-x))
y1_derivative=y1*(1-y1)
plt.plot(x, y1)
plt.plot(x,y1_derivative)
plt.show()
```

### **Relu**

```python
def relu(x):
  x2=[]
  for i in x:
      if i<0:
          x2.append(0)
      else:
          x2.append(i)

  return x2

def relu_derivative(x):
  x1=[]
  for i in x:
      if i<0:
          x1.append(0)
      else:
          x1.append(1)

  return x1


x = np.linspace(-10, 10,100)
y2=relu(x)
y2_derivative=relu_derivative(x)
plt.plot(x,y2 )
plt.plot(x,y2_derivative)
plt.axis('tight')
plt.title('Activation Function :relu')
plt.show()
```

### **tanh**

```python
def tanh(x):

  return np.tanh(x)



x = np.linspace(-10, 10,100)
plt.plot(x, tanh(x),color='red')
plt.plot(x,1-tanh(x)**2)
plt.axis('tight')
plt.title('Activation Function :Tanh')
plt.show()
```

### **leaky_relu**

```python
def leaky_relu(x):
    return max(0.01*x,x)

def der_leaky_relu(x):
    if x < 0 :
        return 0.01
    if x >= 0 :
        return 1
plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: leaky_relu(x),x)), label="leaky-relu")
plt.plot(x, list(map(lambda x: der_leaky_relu(x),x)), label="derivative")
plt.title("Leaky-ReLU")
plt.legend()
plt.show()
```

### **Selu**

```python
def Selu(x):

  landa=1.0507
  alfa=1.67326
  if x<0:
      return (landa*alfa*(np.exp(x) - 1))
  else:
      return landa*x



def der_Selu(x):

  landa=1.0507
  alfa=1.67326
  if x<0:
       return alfa*(np.exp(x) )
  if x>=0:
       return landa

x = np.linspace(-10,10,100)
plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: Selu(x),x)), label="Selu")
plt.plot(x, list(map(lambda x: der_Selu(x),x)), label="derivative")
plt.title("SELU")
plt.legend()
plt.show()



```

### **Elu**

```python
def elu(x):
    if x > 0 :
        return x
    else :
        return (np.exp(x)-1)

def der_elu(x):
    if x > 0 :
        return 1
    else :
        return np.exp(x)
plt.figure(figsize=(12,8))
plt.plot(x, list(map(lambda x: elu(x),x)), label="elu")
plt.plot(x, list(map(lambda x: der_elu(x),x)), label="derivative")
plt.title("ELU")
plt.legend()
plt.show()
```
