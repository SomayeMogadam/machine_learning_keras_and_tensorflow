```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import tensorflow as tf
from tensorflow import keras
  
```

```python
cifar10_dataset= keras.datasets.cifar10
```

```python
(train_images, train_labels),(test_images, test_labels)=keras.datasets.cifar10.load_data()
```

```python
train_images.shape
```

```python
plt.imshow(train_images[1, :, :, :])
```

```python
images = train_images/255.0
images.shape
```

```python
filters = np.zeros((2, 3, 3, 3))

filters[:, :, 0, 0] = 1
filters[:, :, 1, 1] = 1
filters[:, :, 2, 2] = 1

print(filters)
```

```python
plt.imshow(filters[0, :, :, :])
```

```python
convolved_images = tf.nn.conv2d(images[:10], filters, strides=1, padding="SAME")
```

```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Display 10 images
plt.figure(figsize=(15, 6))
for i in range(10):
    plt.subplot(4, 10, i+1 )
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[i][0]])
    
    
    plt.subplot(4, 10, i + 11)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(convolved_images[i], cmap=plt.cm.binary)
plt.show()
```
